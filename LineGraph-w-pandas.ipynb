### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
plt.title('Gas Prices over time (in USD)')

# First way

plt.plot(gas.Year, gas.Canada,'r.-',label='Canada')
plt.plot(gas.Year, gas.USA, 'b.-',label='USA')
plt.plot(gas.Year, gas.Australia, 'y.-',label='Australia')
plt.plot(gas.Year, gas.France, 'k.-',label='France')
plt.plot(gas.Year, gas.Italy, 'c.-',label='Italy')
plt.plot(gas.Year, gas.UK, 'm.-',label='UK')
plt.plot(gas.Year, gas['South Korea'], 'g.-',label='South Korea')

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('Dollars')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# Second way but specific countries

plt.title('Gas Prices over time (Second way but specific countries)')

countries_to_look_at=['Canada','USA','Australia','France']
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker='.',label=country)

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('Dollars')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# Second way

plt.title('Gas Prices over time (Second way All countries)')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.',label=country)

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('Dollars')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
