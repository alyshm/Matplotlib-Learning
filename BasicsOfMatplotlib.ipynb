### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. when saving prolly use 300dpi if possible)
plt.figure(figsize=(5,3),dpi=100)

# Line 1

# Keyword Arguement Notation
# plt.plot(x, y, label='2x', color='blue', linewidth=2, marker='.', markersize=10, markeredgecolor='white')

# Shorthand Notation
# fnt = '[color][marker][line]'
plt.plot(x, y, 'b^--' ,label='2x')

## Line 2

# Select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as Line
plt.plot(x2[:6],x2[:6]**2, color='r', label='x^2')

# Plot remainder of graph as a dot
plt.plot(x2[4:],x2[4:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!',fontdict = {'fontname': 'Comic sans MS','fontsize': '20'})

# X and Y Labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X and Y Axis TickMarks (scaling of your graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

# Save figure (300dpi is good and recommanded when saving so graph has high resolution)
plt.savefig('Ourfirstgraph.png',dpi=300)

# Add a Legend
plt.legend()

# show plot and igore header default line
plt.show()
```

### Bar Charts

```python
labels = ['A', 'B', 'C']
Values = [8,4,6]

bars = plt.bar(labels,Values)

patterns = ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('o')
# bars[1].set_hatch('*')
# bars[2].set_hatch('/')

plt.figure(figsize=(6,4))

plt.show()
```
